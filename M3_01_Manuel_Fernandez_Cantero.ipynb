```python
#Ejercicio primero de POO
#Clase alumno con atributos
class Alumno:
    def __init__(self, nombre, nota):
        self.nombre = nombre
        self.nota = nota
        print('Alumno creado correctamente')

    #Calificacion con if y else para aprobar o suspender
    def calificacion(self):
        if self.nota >= 5:
            print("Aprobado")
        else:
            print("Suspendido")

Lolito = Alumno('Lolito', 9)
Clara = Alumno('Clara', 4)
Clara.calificacion()
Lolito.calificacion()
```

```python
#Ejercicio 2, añadimos __str__

class Alumno:
    def __init__(self, nombre, nota):
        self.nombre = nombre
        self.nota = nota
        print('Alumno creado correctamente')

    #Calificacion con if y else para aprobar o suspender
    def calificacion(self):
        if self.nota >= 5:
            print("Aprobado")
        else:
            print("Suspendido")
    
    def __str__(self):
        return 'Alumno/a: {}; Nota: {}'.format(self.nombre, self.nota)


Lolito = Alumno('Lolito', 9)
print(Lolito)
Clara = Alumno('Clara', 4)
print(Clara)
```

```python
#Ejercicio 3, la clase producto
class Producto:
    def __init__(self, codigo, nombre, precio, tipo):
        self.codigo = codigo
        self.nombre = nombre
        self.precio = precio
        self.tipo = tipo
        print('Producto creado')
    def __str__(self):
        return """
        Código: {}
        Nombre: {}
        Precio: {}
        Tipo:   {}""".format(self.codigo, self.nombre, self.precio, self.tipo)

P1 = Producto('93749347', 'Silla del Don José', 89, 'q?')
print(P1)
```

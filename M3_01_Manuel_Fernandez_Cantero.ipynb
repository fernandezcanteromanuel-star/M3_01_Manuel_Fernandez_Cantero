```python
#Ejercicio primero de POO
#Clase alumno con atributos
class Alumno:
    def __init__(self, nombre, nota):
        self.nombre = nombre
        self.nota = nota
        print('Alumno creado correctamente')

    #Calificacion con if y else para aprobar o suspender
    def calificacion(self):
        if self.nota >= 5:
            print("Aprobado")
        else:
            print("Suspendido")

Lolito = Alumno('Lolito', 9)
Clara = Alumno('Clara', 4)
Clara.calificacion()
Lolito.calificacion()
```

```python
#Ejercicio 2, aÃ±adimos __str__

class Alumno:
    def __init__(self, nombre, nota):
        self.nombre = nombre
        self.nota = nota
        print('Alumno creado correctamente')

    #Calificacion con if y else para aprobar o suspender
    def calificacion(self):
        if self.nota >= 5:
            print("Aprobado")
        else:
            print("Suspendido")
    
    def __str__(self):
        return 'Alumno/a: {}; Nota: {}'.format(self.nombre, self.nota)


Lolito = Alumno('Lolito', 9)
print(Lolito)
Clara = Alumno('Clara', 4)
print(Clara)
```
